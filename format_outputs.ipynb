```python
import pandas as pd

def summarize_for_latex_table(df):
    summary = (
        df
        .groupby(["policy", "environment"])
        .agg(
            avg_first_arrival=("first_arrival_time", "mean"),
            median_first_arrival=("first_arrival_time", "median"),
            success_rate=("success", lambda x: 100 * x.mean()),
            avg_alerts=("num_alerted", "mean"),
            max_alerts=("num_alerted", "max"),
        )
        .reset_index()
    )

    # Optional: round for presentation
    summary["avg_first_arrival"] = summary["avg_first_arrival"].round(2)
    summary["median_first_arrival"] = summary["median_first_arrival"].round(2)
    summary["success_rate"] = summary["success_rate"].round(1)
    summary["avg_alerts"] = summary["avg_alerts"].round(2)

    return summary
```

```python
table_df = summarize_for_latex_table(pd.read_csv('results\combined_results_20260126_173531.csv'))
table_df
```
